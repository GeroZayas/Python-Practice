```python
# quick sort algo


def quick_sort(arr_words: list) -> list:
    if len(arr_words) <= 1:
        return arr_words

    pivot = arr_words[0]

    left = []
    right = []

    for word in arr_words[1:]:
        if len(word) < len(pivot):
            left.append(word)
        else:
            right.append(word)

    return quick_sort(left) + [pivot] + quick_sort(right)
```

```python
quick_sort(words)
```

```python
insertion_sort(words) == quick_sort(words)
```

```python
# merge sort algo


def merge_sort(arr_words: list) -> list:
    if len(arr_words) <= 1:
        return arr_words

    mid = len(arr_words) // 2

    left = arr_words[:mid]
    right = arr_words[mid:]

    left = merge_sort(left)
    right = merge_sort(right)

    return merge(left, right)


def merge(left: list, right: list) -> list:
    ord_arr_words = []
    left_index = right_index = 0

    while left_index < len(left) and right_index < len(right):
        if len(left[left_index]) < len(right[right_index]):
            ord_arr_words.append(left[left_index])
            left_index += 1
        else:
            ord_arr_words.append(right[right_index])
            right_index += 1

    ord_arr_words += left[left_index:]
    ord_arr_words += right[right_index:]

    return ord_arr_words
```
