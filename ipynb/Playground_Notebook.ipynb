```python
# Formatter functions (no classes)
def format_csv(data):
    result = "Name,Age,City\n"
    for item in data:
        result += f"{item['name']},{item['age']},{item['city']}\n"
    return result


def format_html(data):
    result = "<table><tr><th>Name</th><th>Age</th><th>City</th></tr>"
    for item in data:
        result += f"<tr><td>{item['name']}</td><td>{item['age']}</td><td>{item['city']}</td></tr>"
    return result + "</table>"


def format_json(data):
    import json

    return json.dumps(data, indent=2)


# Simple function that uses formatters
def generate_report(data, formatter):
    return formatter(data)


# Usage
data = [
    {"name": "Alice", "age": 30, "city": "NYC"},
    {"name": "Bob", "age": 25, "city": "LA"},
]


def sep():
    print("-" * 80)


# Choose formatter at runtime
sep()
print(generate_report(data, format_csv))
sep()
print(generate_report(data, format_html))
sep()
print(generate_report(data, format_json))
```
