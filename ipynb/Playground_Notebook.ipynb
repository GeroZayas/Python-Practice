```python
import numpy as np
```

```python
x = np.array(range(1,6))
```

```python
x
```

# 0-1 Normalization

```python
scaled = (x - x.min()) / (x.max() - x.min())
```

```python
scaled
```

# Standardization - Z Score
> Centers at 0, std = 1

```python
x.std()
```

```python
x.mean()
```

```python
z = (x - x.mean()) / x.std()
```

```python
z
```
